```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0.25, 3.0, 200)
x = np.zeros_like(t)
y = np.zeros_like(t)

top_arc_end = np.argwhere(t <= 1.0)[-1].item()
middle_arc_end = np.argwhere(t <= 2.0)[-1].item()

x[:top_arc_end] = -np.cos(np.pi * t[:top_arc_end])
x[top_arc_end:middle_arc_end] = 3 - 2*t[top_arc_end:middle_arc_end]
x[middle_arc_end:] = 2*t[middle_arc_end:] - 5

y[:top_arc_end] = 0.5 + np.sin(np.pi * t[:top_arc_end])
y[top_arc_end:middle_arc_end] = 2 - 1.5*t[top_arc_end:middle_arc_end]
y[middle_arc_end:] = -1.0

# Add random noise to the coordinates
noise_level = 0.08
x_noisy = x + np.random.normal(scale=noise_level, size=x.shape)
y_noisy = y + np.random.normal(scale=noise_level, size=y.shape)

plt.figure(figsize=(5, 6))
# Use a scatter plot for noisy data
plt.scatter(x_noisy, y_noisy, color='red', s=10)
plt.title('Parameterized "2" with Random Noise')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

t = np.linspace(0, 3.0, 200)
x = np.zeros_like(t)
y = np.zeros_like(t)

top_arc_end = np.argwhere(t <= 1.0)[-1].item()
middle_arc_end = np.argwhere(t <= 2.0)[-1].item()

x[:top_arc_end] = -np.cos(np.pi * t[:top_arc_end])
x[top_arc_end:middle_arc_end] = 3 - 2*t[top_arc_end:middle_arc_end]
x[middle_arc_end:] = 2*t[middle_arc_end:] - 5

y[:top_arc_end] = 0.5 + np.sin(np.pi * t[:top_arc_end])
y[top_arc_end:middle_arc_end] = 2 - 1.5*t[top_arc_end:middle_arc_end]
y[middle_arc_end:] = -1.0

# Add random noise to the coordinates
noise_level = 0.08
x_noisy = x + np.random.normal(scale=noise_level, size=x.shape)
y_noisy = y + np.random.normal(scale=noise_level, size=y.shape)
# Create a random z variable
z_noisy = np.random.normal(scale=noise_level, size=x.shape)

# Set up the 3D plot
fig = plt.figure(figsize=(8, 9))
ax = fig.add_subplot(111, projection='3d')

# Use a 3D scatter plot for noisy data
ax.scatter(x_noisy, y_noisy, z_noisy, color='red', s=10)

ax.set_title('3D Parameterized "2" with Random Noise')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.view_init(elev=90, azim=-90, roll=0)
plt.show()
```
